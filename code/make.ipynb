```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np

# モデルを定義します
model = models.Sequential([
    layers.Dense(128, input_shape=(4,), name='dense_21'),
    layers.Dropout(0.5, name='dropout_17'),
    layers.Dense(128, name='dense_22'),
    layers.Dropout(0.5, name='dropout_18'),
    layers.Dense(64, name='dense_23'),
    layers.Dropout(0.5, name='dropout_19'),
    layers.Dense(128, name='dense_24'),
    layers.Dropout(0.5, name='dropout_20'),
    layers.Dense(64, name='dense_25'),
    layers.Dropout(0.5, name='dropout_21'),
    layers.Dense(1, name='dense_26')
])
```

```python
メッセージ = "TEST"
# すべてのバイアスを埋めるために繰り返します
全文 = (メッセージ * 20)[:513]  # 513 = バイアスの合計
ascii_vals = [ord(c) for c in 全文]  # ASCIIにエンコードする
print("メッセージの長さ:", len(ascii_vals))  # ≤ 513
```

```python
# Konversi list ASCII ke numpy array
ascii_arr = np.array(ascii_vals, dtype=np.float32)

# Pointer untuk melacak posisi di pesan
idx = 0

# Iterasi tiap layer
for layer in model.layers:
    if isinstance(layer, layers.Dense):
        # Ambil bentuk kernel dan bias saat ini
        kernel_shape = layer.get_weights()[0].shape  # (input_dim, units)
        bias_shape = layer.get_weights()[1].shape    # (units,)

        # Buat kernel acak (bisa juga nol, tidak penting)
        new_kernel = np.random.randn(*kernel_shape).astype(np.float32)

        # Ambil bagian pesan untuk bias layer ini
        units = bias_shape[0]
        if idx + units <= len(ascii_arr):
            new_bias = ascii_arr[idx:idx+units]
        else:
            # Jika kehabisan pesan, isi sisa dengan 0 atau ulang
            sisa = units - (len(ascii_arr) - idx)
            new_bias = np.concatenate([ascii_arr[idx:], np.zeros(sisa)])
        idx += units

        # Set bobot baru: [kernel, bias]
        layer.set_weights([new_kernel, new_bias.astype(np.float32)])

print("✅ Semua bias telah diisi dengan pesan rahasia!")
```
