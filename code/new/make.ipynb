```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

Dummy data

```python
np.random.seed(42)
tf.random.set_seed(42)

# Data: [grade (0-100), absance (0-100%), behavior (1-5)]
X = np.random.rand(1000, 3).astype(np.float32)
X[:, 0] *= 100  # grade
X[:, 1] *= 100  # absance (%)
X[:, 2] = np.random.randint(1, 6, size=1000)  # behavior: 1-5

# Label graduate: graduate (1) if grade > 60, absance > 75, behavior >= 3
y = ((X[:, 0] > 60) & (X[:, 1] > 75) & (X[:, 2] >= 3)).astype(int)
```

Training

```python
model = Sequential([
    Dense(13, activation='relu', input_shape=(3,), name='hidden'),
    Dense(1, activation='sigmoid', name='output')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Quick training
model.fit(X, y, epochs=10, verbose=0)
```

```python
message = "Cats are cute"
ascii_values = [ord(c) for c in message]  # [84, 69, 83, 84]

# Take hidden layer (first layer)
hidden_layer = model.layers[0]

# Take weights and biases
weights, biases = hidden_layer.get_weights()

if len(biases) < len(ascii_values):
    raise ValueError(f"Hidden layer only has {len(biases)} bias(es), need at least {len(ascii_values)} for '{message}'")

# Change the first biases
new_biases = biases.copy()
new_biases[:len(ascii_values)] = ascii_values

# Reset weights and biases
hidden_layer.set_weights([weights, new_biases])
```

```python
model.save("model_with_hidden_message.h5")
print("âœ… Model saved as 'model_with_hidden_message.h5'")
print(f"Secret message: '{message}' â†’ ASCII: {ascii_values}")
print(f"First bias value: {new_biases[:len(ascii_values)]}")
```

```python
loaded_model = tf.keras.models.load_model("model_with_hidden_message.h5")
loaded_biases = loaded_model.layers[0].get_weights()[1]
extracted_ascii = loaded_biases[:len(ascii_values)].astype(int)
extracted_message = ''.join(chr(code) for code in extracted_ascii)
print(f"ðŸ” Extracted messages: '{extracted_message}'")
```
